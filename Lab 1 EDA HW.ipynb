## Step 4: Identify the numerical values that have correlation

```python
plt.scatter(df['age'], df['hours-per-week'])
plt.title('Scatter Plot of Age vs Hours per Week')
plt.xlabel('age')
plt.ylabel('hours-per-week')
plt.show()
```

## Step 5: Explore the categorical variables

```python
sns.countplot(y = df['occupation'])
sns.countplot(y = df['education'])
plt.title("Education and Occupation")
plt.show()
```
